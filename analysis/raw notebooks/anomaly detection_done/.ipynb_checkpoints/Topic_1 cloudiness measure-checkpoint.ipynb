## Write a function that takes as input the dataframe of the weather data, and a variable name (this can be irradiation or module_temperature when calling the function)
#### The function uses this variable to derive the C_day in the following way
Removes outliers (by taking mean+2std


calculates the max curve using the cleaned data


the function returns the array of C_days in each dataframe raw 


```python
import pandas as pd
from datetime import datetime, timedelta
from matplotlib import pyplot as plt
from matplotlib import dates as mpl_dates
import seaborn as sns
import numpy as np
```

```python
weather_data = pd.read_csv('Plant_1_Weather_Sensor_Data.csv')
```

```python
def find_cloudiness(data, col_name):
    data['DATE_TIME']= pd.to_datetime(data['DATE_TIME'])
    data['DAY'] = pd.DatetimeIndex(data['DATE_TIME']).dayofyear
    data['TIME'] = data.DATE_TIME.dt.time
    data['HOUR'] = data.DATE_TIME.dt.hour
    agg_h_irr = data.groupby(data.TIME).agg({col_name: 'max'})
    #Removing outliers 
    clean_data=data.copy()
    for i, j in  clean_data.groupby(clean_data.HOUR):
        outlier_condition1=(j[col_name]!=0)&(j[col_name]>=(j[col_name].mean()+ 2*j[col_name].std()))
        outlier_condition2=(j[col_name]!=0)&(j[col_name]<(j[col_name].mean()- 2*j[col_name].std()))
        clean_data.drop(clean_data.loc[(clean_data['DAY'].isin(j.loc[outlier_condition1|outlier_condition2].DAY))&
                                       (clean_data['HOUR']==i)].index, inplace=True)
    agg_h_irr_clean = clean_data.groupby(clean_data.TIME).agg({col_name: 'max'})
    weather_maxirr = pd.merge(data, agg_h_irr_clean, how='inner', on='TIME', suffixes=('', '_max'))
    weather_maxirr['offset_from_max']=(weather_maxirr[col_name+'_max']-weather_maxirr[col_name])**2
    C_day_list=weather_maxirr.groupby('DAY').sum()['offset_from_max']
    weather_maxirr = pd.merge(weather_maxirr, C_day_list, how='inner', on='DAY')
    weather_maxirr_new = weather_maxirr.rename(columns={'offset_from_max_y': 'C_day', 
                                                        'offset_from_max_x': 'offset_from_max'})
    return(weather_maxirr_new[col_name])
```

```python
find_cloudiness(weather_data, 'IRRADIATION')
```

```python
find_cloudiness(weather_data, 'MODULE_TEMPERATURE')
```

##### Checking if the C_day calculated using irradiation and the one calculated using module temperature are similar

```python
plt.figure(figsize=(15, 10))
scale = find_cloudiness(weather_data, 'IRRADIATION').max()*2 / find_cloudiness(weather_data, 'MODULE_TEMPERATURE').max()
plt.plot(find_cloudiness(weather_data, 'IRRADIATION'))
plt.plot(find_cloudiness(weather_data, 'MODULE_TEMPERATURE')*scale, color='red')
plt.title('Cloudiness measures using Module temperature(red) and irradiation')
plt.show()
```
