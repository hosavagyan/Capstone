# Finalizing plots

```python
import pandas as pd
from datetime import datetime, timedelta
from matplotlib import pyplot as plt
from matplotlib import dates as mpl_dates
import seaborn as sns
import numpy as np
```

```python
weather_data = pd.read_csv('Plant_1_Weather_Sensor_Data.csv')
gen_data = pd.read_csv('Plant_1_Generation_Data.csv')
```

```python
weather_data['DATE_TIME']= pd.to_datetime(weather_data['DATE_TIME'])
gen_data['DATE_TIME']= pd.to_datetime(gen_data['DATE_TIME'], format='%d-%m-%Y %H:%M')
```

```python
weather_data.head()
```

```python
gen_data.head()
```

```python
# Daily maximum irradiation, module and ambient temp
agg_df = weather_data.set_index('DATE_TIME').resample('D').agg({'IRRADIATION': 'max',
                                                                'MODULE_TEMPERATURE': 'max',
                                                               'AMBIENT_TEMPERATURE': 'max'})
agg_df['DAY'] = pd.DatetimeIndex(agg_df.index).dayofyear
agg_df.head()
```

```python
weather_data['DAY'] = pd.DatetimeIndex(weather_data['DATE_TIME']).dayofyear
weather_data['TIME'] = weather_data.DATE_TIME.dt.time
weather_data['HOUR'] = weather_data.DATE_TIME.dt.hour
weather_data.head()
```

```python
#Hourly maximum irradiation, module and ambient temp
agg_h_irr = weather_data.groupby(weather_data.DATE_TIME.dt.hour).agg({'IRRADIATION': 'max',
                                                            'MODULE_TEMPERATURE': 'max',
                                                            'AMBIENT_TEMPERATURE': 'max'})
agg_h_irr.head()
```

```python
merged_df = pd.merge(gen_data, weather_data, how='inner', on='DATE_TIME')
merged_df
```

## Overplot irradiation profiles for many days

Irradiation profile of each day

```python
for i, day in weather_data.groupby(weather_data['DATE_TIME'].dt.dayofyear):
    plt.figure(figsize=(15, 10))
    plt.plot_date(day['DATE_TIME'], day['IRRADIATION'])
    plt.title(datetime(2020, 1, 1) + timedelta(int(i) - 1))
    plt.ylabel('Irradiation')
    plt.show()
```

```python
plt.figure(figsize=(15, 10))
plt.plot_date(agg_df.index, agg_df['IRRADIATION'])
plt.title("Daily maximum irradiation")
plt.ylabel('Irradiation')
plt.xlabel('Days')
plt.show()
```

Hourly irradiation grouped by days, and hourly maximum irradiation

```python
plt.figure(figsize=(15, 10))
for a, b in weather_data.groupby(weather_data['DAY']):
    plt.scatter(b['HOUR'], b['IRRADIATION'])
plt.scatter(agg_h_irr.index, agg_h_irr['IRRADIATION'], color='black')
plt.legend(agg_df['DAY'])
plt.ylabel('Irradiation')
plt.xlabel('Hours')
plt.show()
```

Everyday hourly irradiation and hourly maximum irradiation

```python
for a, b in weather_data.groupby(weather_data['DAY']):
    plt.figure(figsize=(15, 10))
    plt.scatter(b['HOUR'], b['IRRADIATION'])
    plt.scatter(agg_h_irr.index, agg_h_irr['IRRADIATION'], color='red')
    plt.title(datetime(2020, 1, 1) + timedelta(int(a) - 1))
    plt.ylabel('Irradiation')
    plt.xlabel('Hours')
    plt.show()
```

# Correlations

## Correlation coefficient of irradiation and module temperature for each day. You will have an array of 30 coefficients for 30 days. Plot the distribution of these coefficients

```python
mod_irr = []
for a, b in weather_data.groupby(weather_data['DAY']):
    c = b['IRRADIATION'].corr(b['MODULE_TEMPERATURE'])
    mod_irr.append(c)

mod_irr = pd.DataFrame(data=np.array(mod_irr), columns=['CORRELATION'])
mod_irr['DAY'] = weather_data['DAY'].unique()
mod_irr.head()
```

```python
plt.figure(figsize=(15, 10))
plt.scatter(mod_irr['DAY'], mod_irr['CORRELATION'])
plt.title('Distribution of correlation coefficients of irradiation and module temperature for each day')
plt.ylabel('Correlation coefficient')
plt.xlabel('Days')
plt.show()
```

## Calculate correlation of irradiation and DC power for each day. 
In this case you will have 30xn_inv coefficients, where n_inv is the number of inverters in the plant. Try to plot the values against the day. So for each day you will have a set of points that show the correlation of DC power of a set of converters with the irradiation



```python
df_corr = pd.DataFrame(columns=['DAY', 'SOURCE_KEY', 'IRR_DC_CORR'])
i = 0
for _, by_day in merged_df.groupby(by=merged_df['DATE_TIME'].dt.dayofyear):
    for __, by_source in by_day.groupby(by='SOURCE_KEY_x'):
        df_corr.loc[i] = [datetime(2020, 1, 1) + timedelta(int(_) - 1) ,__,
                          by_source['IRRADIATION'].corr(by_source['DC_POWER'])]
        i+=1
df_corr                   
```

```python
plt.figure(figsize=(15, 10))
for a, b in df_corr.groupby(df_corr['SOURCE_KEY']):
    plt.scatter(b['DAY'], b['IRR_DC_CORR'])
plt.legend(df_corr['SOURCE_KEY'])
plt.ylabel('correlation of irradiation and DC power')
plt.xlim(datetime(2020, 5, 14, 0, 0), datetime(2020, 6, 18, 0, 0))
plt.show()
```
